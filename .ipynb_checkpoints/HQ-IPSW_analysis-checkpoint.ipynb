```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#!ls python_projects/IPSW/HQ-IPSW/Data_18months_Outlook/*/0*.xlsx

path = "Data_18months_Outlook/*/*.xlsx"

import glob
Full_path = glob.glob(path)
print(np.shape(Full_path))
#glob.glob(path)

all_data = pd.DataFrame()
for f in glob.glob(path):
    df = pd.read_excel(f)
    all_data = all_data.append(df,ignore_index=True)
    
# convert the date column to a date time object
all_data['Date (week ending)'] = pd.to_datetime(all_data['Date (week ending)'])

# Description of data
all_data.describe()

latest_ind = list(np.array([0, 7, 9, 12, 13]))    #14
older_ind = list(np.array([1,2,3,4,5,6,8,10,11,14,15,16,17,18,19,20,21]))
myorder = older_ind + latest_ind

mylist = [glob.glob(path)[i] for i in myorder]

#train_data = pd.DataFrame()
#for f in glob.glob(path)[latest_ind]:
#    df = pd.read_excel(f)
#    train_data = train_data.append(df,ignore_index=True)
#    
#test_data = pd.DataFrame()
#for f in glob.glob(path)[latest_ind]:
#    df = pd.read_excel(f)
#    test_data = test_data.append(df,ignore_index=True)

    
train_data = pd.DataFrame()
for f in mylist[0:21]:
    df = pd.read_excel(f)
    train_data = train_data.append(df,ignore_index=True)
    
test_data = pd.DataFrame()
for f in mylist[21:]:
    df = pd.read_excel(f)
    test_data = test_data.append(df,ignore_index=True)

train_data['Date (week ending)'] = pd.to_datetime(train_data['Date (week ending)'])
test_data['Date (week ending)'] = pd.to_datetime(test_data['Date (week ending)'])

print(np.shape(test_data))
```

```python
# Drop rows with missing values in selected columns

# all data
df_all = all_data.dropna(axis = 0, how='any', subset=['HOEP'])
df_all = df_all.dropna(axis = 0, how='any', subset=['Baseload Generation'])

# test data
df_test = test_data.dropna(axis = 0, how='any', subset=['East PD','Baseload Generation'])
#df_test = df_test.dropna(axis = 0, how='any', subset=['Baseload Generation'])

# train data
df = train_data.dropna(axis = 0, how='any', subset=['HOEP'])
df = df.dropna(axis = 0, how='any', subset=['Baseload Generation'])

df_test.describe()
```

```python
# Pearson Correlation analysis

from scipy.stats import pearsonr

K_corr = 10 # Number of selected features with correlation
corr_list = []
hoep = df['HOEP']
columns_name = df.columns.ravel()


for i in columns_name[2:-1]:
    param = df[i]
    #print(param[0:5].T)
    nans = np.isnan(np.array(param))
    #print(hoep[~nans][1:5])
    corr, _ = pearsonr(hoep[~nans], param[~nans])
    corr_list.append(np.abs(corr))
    
#print(np.abs(corr_list))
corr_dictionary = dict(zip(columns_name[2:-1], corr_list))

import operator
sorted_corr = {k: v for k, v in sorted(corr_dictionary.items(), key=lambda item: item[1], reverse=True)}
#print(sorted_corr)

#print(len(columns_name))
corr_ind_dictionary = dict(zip(np.add(range(len(columns_name)-2),2), corr_list))
sorted_ind_corr = {k: v for k, v in sorted(corr_dictionary.items(), key=lambda item: item[1], reverse=True)}
inds_sorted = list(sorted_ind_corr.keys()) 

# Selected features with MI
sel_feat_corr = inds_sorted[0:K_corr]
print(sel_feat_corr)
print(['Selected features with Pearson Correlation:', sel_feat_corr])


# Plot the correlations
keys = list(sorted_corr.keys()) 
values = list(sorted_corr.values()) 
   
fig = plt.figure(figsize = (10, 5)) 
  
# creating the bar plot 
plt.bar(keys, values, color ='maroon') 
plt.xlabel("Variables") 
plt.xticks(rotation=90)
plt.ylabel("Pearson correlation coefficient") 
plt.title("Correlation of each parameter to HOEP") 
plt.savefig('Figures/Correlation.png',bbox_inches='tight')
plt.show() 
```

```python
# Mutual Information
K_mi = 10 # Number of features selected with MI

import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_selection import f_regression, mutual_info_regression

X = np.array(df[0:-1][keys])
print(X.shape)
Y = df[0:-1]['HOEP']

f_test, _ = f_regression(X, Y)
f_test /= np.max(f_test)
mi = mutual_info_regression(X, Y)
mi /= np.max(mi)



ind_sort = np.argsort(-mi)
mi_dictionary = dict(zip(columns_name[2:-1], mi))

import operator
sorted_mi = {k: v for k, v in sorted(mi_dictionary.items(), key=lambda item: item[1], reverse=True)}
#print(sorted_mi)

#print(len(columns_name))
mi_ind_dictionary = dict(zip(np.add(range(len(columns_name)-2),2), corr_list))
sorted_ind_mi = {k: v for k, v in sorted(mi_dictionary.items(), key=lambda item: item[1], reverse=True)}
inds_sorted = list(sorted_ind_mi.keys()) 

# Selected features with MI
sel_feat_mi = inds_sorted[0:K_mi]
print(['Selected features with MI:', sel_feat_mi])

# Plot the correlations
keys = list(sorted_mi.keys()) 
values = list(sorted_mi.values()) 
   
fig = plt.figure(figsize = (10, 5)) 
  
# creating the bar plot 
plt.bar(keys, values, color ='indigo') 
plt.xlabel("Variables") 
plt.xticks(rotation=90)
plt.ylabel("MI") 
plt.title("Mutual information of HOEP and variables") 
plt.savefig('Figures/MI.png',bbox_inches='tight')
plt.show() 

plt.figure(figsize=(15, 5))
K = 4
for i in range(K):
    plt.subplot(1,K, i + 1)
    plt.scatter(X[:, ind_sort[i]], Y, edgecolor='black', s=20)
    plt.xlabel("$x_{}$".format(i + 1), fontsize=14)
    if i == 0:
        plt.ylabel("$y$", fontsize=14)
    plt.title("MI={:.2f}".format(mi[ind_sort[i]]),
              fontsize=16)
plt.show()
```

```python
# Feature selection using regularization with LASSO and regression

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, LogisticRegression, LassoCV
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error


# Train test spilit
X_train_full, X_val_full, y_train, y_val = train_test_split(
    df.drop(labels=['HOEP'], axis=1),
    df['HOEP'],
    test_size=0.25,
    random_state=0)


X_train = X_train_full.drop(labels=['Date (week ending)'], axis=1)
X_val = X_val_full.drop(labels=['Date (week ending)'], axis=1)

X_test = df_test[columns_name[2:]]
y_test = df_test['HOEP']

X_test = np.array(X_test)

#X_train = X[0:-300,:]
#y_train = Y[0:-300]

scaler = StandardScaler()
scaler.fit(X_train)

columns_name = df.columns.ravel()
Features_name = columns_name[2:]

# here, again I will train a Lasso Linear regression and select
# the non zero features in one line.
# bear in mind that the linear regression object from sklearn does
# not allow for regularisation. So If you want to make a regularised
# linear regression you need to import specifically "Lasso"
# that is the l1 version of the linear regression
# alpha is the penalisation here, so I set it high in order
# to force the algorithm to shrink some coefficients

# Regression task
sel_ = SelectFromModel(Lasso(alpha=.3))
sel_.fit(scaler.transform(X_train), y_train)

# Classification task
#sel_ = SelectFromModel(LogisticRegression(C=1, penalty='l1'))
#sel_.fit(scaler.transform(X_train), y_train)

sel_.get_support()

# make a list with the selected features and print the outputs
selected_feat = Features_name[(sel_.get_support())]

print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(sel_.estimator_.coef_ == 0)))

sel_feat_index = np.array(sel_.get_support())
#print(Features_name[(sel_.get_support())])

sel_feat_lasso = Features_name[(sel_.get_support())]
print(sel_feat_lasso)
```

```python
# Gradient boosting
from sklearn import datasets, ensemble

params = {'n_estimators': 500,
          'max_depth': 4,
          'min_samples_split': 5,
          'learning_rate': 0.01,
          'loss': 'ls'}
reg = ensemble.GradientBoostingRegressor(**params)
reg.fit(X_train, y_train)

rmse = np.sqrt(mean_squared_error(y_val, reg.predict(X_val)))
print("The mean squared error (MSE) on test set: {:.4f}".format(rmse))


test_score = np.zeros((params['n_estimators'],), dtype=np.float64)
for i, y_pred in enumerate(reg.staged_predict(X_val)):
    test_score[i] = reg.loss_(y_val, y_pred)

fig = plt.figure(figsize=(6, 6))
plt.subplot(1, 1, 1)
plt.title('Deviance')
plt.plot(np.arange(params['n_estimators']) + 1, reg.train_score_, 'b-',
         label='Training Set Deviance')
plt.plot(np.arange(params['n_estimators']) + 1, test_score, 'r-',
         label='Test Set Deviance')
plt.legend(loc='upper right')
plt.xlabel('Boosting Iterations')
plt.ylabel('Deviance')
fig.tight_layout()
plt.show()


```

```python
from sklearn.inspection import permutation_importance

feature_importance = reg.feature_importances_
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
fig = plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, np.array(columns_name)[sorted_idx+2])
plt.title('Feature Importance (MDI)')

result = permutation_importance(reg, X_val, y_val, n_repeats=10,
                                random_state=42, n_jobs=2)
sorted_idx = result.importances_mean.argsort()
plt.subplot(1, 2, 2)
plt.boxplot(result.importances[sorted_idx].T,
            vert=False, labels=np.array(columns_name)[sorted_idx+2])
plt.title("Permutation Importance (test set)")
fig.tight_layout()
plt.show()


```

```python
# Estimating monthly cost to compare with bench mark

from datetime import datetime, timedelta

def estimate_monthly_cost(weekly_HOEP, weekly_datetime):
    
    weekly_dict = dict(zip(weekly_datetime, weekly_HOEP))
    
    daily_dict = {}
    for key in weekly_dict:
        for days_to_subtract in range(7):
            new_key = key - timedelta(days=days_to_subtract)
            #daily_dict[new_key] = weekly_dict[key]
            daily_dict.update({new_key: weekly_dict[key]})
    
    #print(list(daily_dict.keys()))
    output_df = []
    output_df = pd.DataFrame()
    output_df['Date'] = list(daily_dict.keys())
    output_df['HOEP'] = list(daily_dict.values())
    output_df['Date'] = pd.to_datetime(output_df['Date'])
    #print(output_df.head)
    return output_df.groupby(output_df.Date.dt.to_period("M")).mean()
```
